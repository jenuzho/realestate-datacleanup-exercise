#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio por metro cuadrado y escribe una conclusión sobre ellos.

Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

```python
# Crea una nueva columna'pps' (price per square/precio x metro cuadrado)
df['pps'] = df['price'] / df['surface']

# Imprime la columna 'pps' creada
print(df[['pps']]) 

# Calcula el promedio de la columna "pps" para Valdemorillo y Galapagar
avg_valdemorillo_pps = round(df[df['level5'] == 'Valdemorillo']['pps'].mean(), 2)
avg_galapagar_pps = round(df[df['level5'] == 'Galapagar']['pps'].mean(), 2)

# Imprime el promedio de la columna 'pps'
# HE AÑADIDO EL SIGNO DEL EURO EN EL STRING PARA FACILITAR LA LECTURA
print(f"PRECIO PROMEDIO del metro cuadrado en VALDEMORRILLO es de: {avg_valdemorillo_pps}€")
print(f"PRECIO PROMEDIO del metro cuadrado en GALAPAGAR es de: {avg_galapagar_pps}€")

# Compara los promedios de la columna "pps" e imprime la conclusión
if avg_valdemorillo_pps == avg_galapagar_pps:
  print("El PRECIO PROMEDIO, por metro cuadrado, entre Valdemorillo y Galapagar ES EL MISMO.")
else:
  print("El PRECIO PROMEDIO, del metro cuadrado, entre Valdemorillo y Galapagar es DIFERENTE.")
  # ... (previous code to calculate avg_valdemorillo_pps and avg_galapagar_pps) ...

# Compara el promedio de la columna pps e imprime el valor más alto
if avg_valdemorillo_pps > avg_galapagar_pps:
    higher_pps = "Valdemorillo"
    percentage_difference = ((avg_valdemorillo_pps - avg_galapagar_pps) / avg_galapagar_pps) * 100
else:
    higher_pps = "Galapagar"
    percentage_difference = ((avg_galapagar_pps - avg_valdemorillo_pps) / avg_valdemorillo_pps) * 100

print(f"El PRECIO PROMEDIO, por metro cuadrado, es más ELEVADO en {higher_pps}.")
print(f"La DIFERENCIA PORCENTUAL es: {percentage_difference:.2f}%")
```

Estos puntos se consideran valores atípicos y pueden representar casas con características especiales que influyen en su precio de manera diferente.

En **(G2)** Los puntos tienden a agruparse en una línea diagonal, que sube de izquierda a derecha, lo que significa que existe una **CORRELACIÓN POSITIVA entre las VARIABLES.** En este caso, indicaría que a medida que aumenta el "Surface Area" (TAMAÑO DE LA CASA), también tiende a aumentar el "Price" (PRECIO). #### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python

# Lista de poblaciones del cinturón sur de Madrid
belt_sur_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Configurar el gráfico
plt.figure(figsize=(10, 6))

# Colores para cada población
colors = ['skyblue', 'salmon', 'lightgreen', 'lightcoral']

# Normalizar los precios para cada población y trazar histogramas
for i, population in enumerate(belt_sur_populations):
    # Filtrar el grupo de la población
    group = df_belt_sur[df_belt_sur['level5'] == population]
    
    # Normalización min-max para los precios de la población
    min_price = group['price'].min()
    max_price = group['price'].max()
    normalized_prices = (group['price'] - min_price) / (max_price - min_price)
    
    # Trazar el histograma de los precios normalizados
    plt.hist(normalized_prices, bins=20, alpha=0.6, color=colors[i], label=population)

# Personalizar el gráfico
plt.title('Histograma de Precios Normalizados en el Cinturón Sur de Madrid')
plt.xlabel('Precios Normalizados (Min-Max)')
plt.ylabel('Frecuencia')
plt.legend(title="Poblaciones")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Crear la nueva columna 'pps' (price per square meter) si no se ha creado ya
df_belt_sur['pps'] = df_belt_sur['price'] / df_belt_sur['surface']

# Poblaciones seleccionadas para el análisis
selected_populations = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']

# Filtrar el DataFrame para estas poblaciones
df_selected = df_belt_sur[df_belt_sur['level5'].isin(selected_populations)]

# Crear subplots (2 filas, 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Ajustar el espacio entre los subplots
fig.tight_layout(pad=5.0)

# Trazar diagramas de dispersión para cada población
for i, population in enumerate(selected_populations):
    # Seleccionar el subplot actual
    ax = axes[i // 2, i % 2]  # Indice en 2x2 subgrids

    # Filtrar los datos para la población actual
    population_data = df_selected[df_selected['level5'] == population]

    # Trazar diagrama de dispersión
    ax.scatter(population_data['surface'], population_data['pps'], alpha=0.6)
    
    # Títulos y etiquetas
    ax.set_title(f'{population}')
    ax.set_xlabel('Superficie (m²)')
    ax.set_ylabel('Precio por m² (pps)')

# Mostrar el gráfico
plt.suptitle('Precio por metro cuadrado vs Superficie en 4 Poblaciones', fontsize=16)
plt.show()
```

```python
# Define las poblaciones del cinturón sur
belt_sur_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filtra el DataFrame para estas poblaciones
df_belt_sur = df[df['level5'].isin(belt_sur_populations)]

# Crea un diccionario para almacenar las coordenadas por población
coordinates_by_population = {}
for population in belt_sur_populations:
    coordinates_by_population[population] = df_belt_sur[df_belt_sur['level5'] == population][['latitude', 'longitude']].values.tolist()

# Traza las coordenadas por color de población
fig, ax = plt.subplots()
for population, coordinates in coordinates_by_population.items():
    latitudes = [coord[0] for coord in coordinates]
    longitudes = [coord[1] for coord in coordinates]
    ax.scatter(longitudes, latitudes, label=population, alpha=0.5)

ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')
ax.set_title('Coordenadas del Cinturón Sur de Madrid')
ax.legend()
plt.show()
```
